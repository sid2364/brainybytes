```python
import numpy as np
# Plotting libraries
import matplotlib.pylab as plt
# Magic function to make matplotlib inline; other style specs must come AFTER
%matplotlib inline
# This enables high resolution PNGs.
%config InlineBackend.figure_formats = {'png', 'svg'}
```

```python
#Payoff matrix prisoners dilemma
T=3; R=7; P=1; S=0
A=np.array([[R,S],[T,P]])
```

```python
from egttools.analytical import StochDynamics
from egttools.plotting import plot_gradients
```

```python
# Parameters and evolver
nb_strategies = 2; Z = 100; N = 2;
beta = 1
pop_states = np.arange(0, Z + 1, 1)
strategy_i = np.linspace(0, 1, num=Z + 1, dtype=np.float64)
evolver = StochDynamics(nb_strategies, A, Z)
```

```python
gradients = np.array([[evolver.gradient_selection(x, 0, 1, beta)]
                      for x in pop_states])
#better alternative gradients = egt.analytical.utils.calculate_gradients(np.array((x, 1 - x)).T, gradient_function)
```

```python
plot_gradients(gradients, figsize=(4,4), fig_title="Dilemma stochastic dynamics",
               marker_facecolor='white',
               xlabel="frequency of C (k/Z)", marker="o", marker_size=20, marker_plot_freq=2)
plt.show()
```
